```python
df_country = pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
#combine country with zomato counry_code col
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

```python
final_df.info()
```

```python
country_name = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
print(final_df.Country.value_counts().index)
print(final_df.Country.value_counts().values)
```

```python
#Pie Chat    -Top 3 countries that uses zomato
plt.pie(country_val[:3], labels=country_name[:3], autopct='%1.2f%%')
```

Observation: zomatos maximum records and trasaction are from India, 4.7% transaction are from US and 0.8% transaction are from UK

```python
final_df.columns
```

```python
rating_grp = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
print(rating_grp)
```

Observation:
1. Rating between 4.5 - 4.9 -----> Excellent
2. Rating between 3.5 - 3.9 -----> Good
3. Rating between 1.8 - 2.4 -----> Poor

```python
sns.barplot(x='Aggregate rating',y='Rating Count', data=rating_grp, hue='Rating color',palette=['white', 'red','orange','yellow','green','green'])
```

Observation:
1. not rated count is very high
2. maximum number of rating are between 2.5 - 3.4

```python
#count plot
sns.countplot(x='Rating color', data=rating_grp, palette=['blue', 'red','orange','yellow','green','green'])
#its based on frequency count
```

```python
#Find the countried name that has given 0 rating
final_df[final_df['Aggregate rating'] == 0.0].groupby('Country').size().reset_index()
```

Observation:
Maimum number of 0 rating is given by Indian customers
